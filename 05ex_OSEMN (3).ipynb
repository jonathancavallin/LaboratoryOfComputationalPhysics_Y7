4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
file_credit_card= "credit_card.dat"
with open(file_credit_card, mode='r') as f4:
    for line in f4:
        #print(line)
        no_space_line = line.strip()[:-4] ##exclude last 4 bits (spaces) f each line
        credit_card_number = ""
        for i in range(0, len(no_space_line),6): ##in each line iterate every 6 every bits
            credit_card_number += chr(int(no_space_line[i:i+6],2)) # Convert 6-bits binary number to decimal, then to character
        print(credit_card_number)
```

5\. Load the file "user_data.json", filter the data by the "CreditCardType" field equals to "American Express". Than save the data to a CSV.

```python
import json
import csv
with open('user_data.json', mode='r') as file:
    data = json.load(file)

#print(data[:201]) #data is a list of dictionaries, each containing keys ID,Job Title,..,CreditCardType

American_Cards = [user for user in data if user['CreditCardType'] == 'American Express'] 
                  #user is a dictionary in the list of dictionaries data
print(American_Cards)
##one way to convert to csv file
with open('American_Cards.csv', mode='w') as f5:
    # Extract column names, the dictionary keys, from the first dictionary of the list of dictionaries
    headers = American_Cards[0].keys()
    
    writer = csv.DictWriter(f5, fieldnames=headers)  # Create a CSV writer
    
    # Write the header row
    writer.writeheader()
    
    # Write the data rows
    writer.writerows(American_Cards)
```

```python
## or we can convert the JSON file American_Cards to csv more easily with pandas !! (learn this)
import pandas as pd

## first create a new JSON file with the filtered users
with open('American_Cards.json', mode= 'w') as f6:
    json.dump(American_Cards, f6) 
    
df = pd.read_json('American_Cards.json') 

df.to_csv('American_Cards_new.csv', index=False) # save the DataFrame to a csv file
```

```python
import pandas as pd

file_name= 'mushrooms_categorized.csv'
mushroom_data = pd.read_csv(file_name) # basically creates a DataFrame 
class_column = mushroom_data['class']  #'class' is a column of the DataFrame
print(class_column)
```

It's a csv file with 8124 rows, corresponding to a different type of mushroom; for each type of mushroom different features/characteristics (bruises, odor, cap color, etc...) are described, either with a discrete number scale or with a 0-1, indicating the presence or absence of the said feature.   

```python
from matplotlib import pyplot as plt

plt.hist(class_column)
plt.title("Histogram of Mushroom Classes")
plt.xlabel("Class")
plt.ylabel("Frequency")
plt.show()
```

```python
capcolor_column = mushroom_data['cap-color']

plt.hist(capcolor_column, color='red', edgecolor='black')
plt.title("Histogram of Mushroom Cap Colour")
plt.xlabel("Cap Colour label")
plt.ylabel("Frequency")
plt.show()
```
